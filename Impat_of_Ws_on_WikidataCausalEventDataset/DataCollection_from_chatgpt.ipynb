```python
from openai import OpenAI
import pandas as pd
import time
import os


client = OpenAI(api_key = "########")
wiki_data_property_df = pd.read_csv("Wikidata_properties_description.csv")
classified_properties = os.listdir("Wikidata_property")
for i in range(0, len(wiki_data_property_df)):
    property_file = str(wiki_data_property_df['id'][i].split("/")[-1]) + '.txt'
    if(property_file in classified_properties):
        continue
    instruction = f"Analyze the Property id and label using the Wikidata property description to determine the most accurate description of its objectives from the following options: \nWhere: possible places. \nWhen: specific time. \nWho: active/passive entities. \nWhy: possible reason of cause.\nHow: method by which an action was performed.\nE: None of the above. \nF: Not sure. \nMention only the selected option in one word"
    request = "Property id: "+wiki_data_property_df['id'][i]+"\nLabel: "+wiki_data_property_df['label'][i]+"\nDescription: "+wiki_data_property_df['description'][i] 

    response = client.chat.completions.create(
        model="gpt-3.5-turbo-1106",
        temperature=0.7,
        messages=[
            {"role": "system", "content": instruction},
            {"role": "user", "content": request}
        ]
    )

    response = response.choices[0].message.content
    with open("Query.txt",'a', encoding='utf-8') as f:
        f.writelines(request)
        f.writelines(instruction)
        f.writelines(response)

    with open('Wikidata_property/' + property_file, 'w', encoding='utf-8') as f:
        f.writelines(response)
    print(i, 'DONE')
    time.sleep(1)
    if(i==12000):
        break
```
